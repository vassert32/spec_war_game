4.3.2 –ó–∞–¥–∞–Ω–∏–µ –î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ.png; –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
from matplotlib import pyplot as plt
import numpy as np
import math

def func(x):
    chisl = np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2))
    znam = np.abs(1 - np.sin(x)**3)
    return chisl / znam

# –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π x –æ—Ç 1 –¥–æ 10 —Å —à–∞–≥–æ–º 0.1
x_values = np.arange(1, 11, 1)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ x
y_values = np.array([func(x) for x in x_values])

# –í—ã–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–∑–∞ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞
half_length = len(x_values) // 2  # –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞ –¥–ª—è –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞
y_slice = y_values[:half_length]
x_slice = x_values[:half_length]

#plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # –ü–µ—Ä–≤—ã–π –≥—Ä–∞—Ñ–∏–∫ –≤ —Ä—è–¥—É –∏–∑ –¥–≤—É—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.plot(x_values, y_values, label='–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤', color='blue')
plt.title('–õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫')


plt.subplot(1, 2, 2)
plt.scatter(x_slice, y_slice, label='–°—Ä–µ–∑ –º–∞—Å—Å–∏–≤–∞', color='red')
plt.title('–¢–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫')

plt.show()

```

–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.

```python
from scipy.integrate import simps
from numpy import trapz
from matplotlib import pyplot as plt
import numpy as np
import math

x = np.arange(0, 11, 1)
y = np.abs(np.cos(x*np.exp(np.cos(x) + np.log(x + 1))))

plt.plot(x, y, c = 'r')
plt.fill_between(x, y)
plt.grid()
plt.show()

area = trapz(y)
print(area)
```

```python
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import datetime

# –°–ø–∏—Å–æ–∫ —Å–æ–∫—Ä–∞—â–µ–Ω–∏–π –¥–ª—è –º–µ—Å—è—Ü–µ–≤
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
dates=[]

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –º–µ–∂–¥—É –º–µ—Å—è—Ü–∞–º–∏
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%b')  # –§–æ—Ä–º–∞—Ç –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ç–æ–ª—å–∫–æ –º–µ—Å—è—Ü–∞

# –ß—Ç–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–æ–≤
apple = open("/content/apple_data.txt").read().splitlines()[1:]
microsoft = open("/content/microsoft_data.txt").read().splitlines()[1:]
google = open("/content/google_data.txt").read().splitlines()[1:]

# –û–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞—Ç –∏ –∞–∫—Ü–∏–π
for i in range(max((len(apple), len(google), len(microsoft)))):
  dates.append(datetime.datetime.strptime(apple[i].split(',')[0], '%Y-%m-%d').date())
  apple[i] = round(float(apple[i].split(',')[4]))
  microsoft[i] = round(float(microsoft[i].split(',')[4]))
  google[i] = round(float(google[i].split(',')[4]))

# –°–æ–∑–¥–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
fig, ax = plt.subplots()

# –£—Å—Ç–∞–Ω–æ–≤–∫–∞ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–µ—Å—Ç–æ–ø–æ–ª–æ–∂–µ–Ω–∏—è –∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –º–µ—Ç–æ–∫ –Ω–∞ –æ—Å–∏ x
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
plt.plot(dates, apple)
plt.plot(dates, microsoft)
plt.plot(dates, google)
plt.legend(['apple','microsoft','google'], loc=2)
# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–∞–¥–ø–∏—Å–∏ –æ –≥–æ–¥–µ –ø–æ–¥ –æ—Å—å—é x
plt.text(0.5, -0.1, '2021', ha='center', transform=ax.transAxes)
plt.show()
```

```python
import math

def calculator(x, y):
    print(f"–°–ª–æ–∂–µ–Ω–∏–µ: {x} + {y} = {x + y}")
    print(f"–í—ã—á–∏—Ç–∞–Ω–∏–µ: {x} - {y} = {x - y}")
    print(f"–£–º–Ω–æ–∂–µ–Ω–∏–µ: {x} * {y} = {x * y}")
    print(f"–î–µ–ª–µ–Ω–∏–µ: {x} / {y} = {x / y if y != 0 else '–û—à–∏–±–∫–∞: –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å'}")
    print(f"–≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞ —Å—É–º–º—ã: e^(x+y) = {math.exp(x + y)}")
    print(f"–°–∏–Ω—É—Å —Å—É–º–º—ã: sin(x + y) = {math.sin(x + y)}")
    print(f"–ö–æ—Å–∏–Ω—É—Å —Å—É–º–º—ã: cos(x + y) = {math.cos(x + y)}")
    print(f"–í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å: x^y = {x ** y}")

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä–∞
x = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ x: "))
y = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ y: "))

calculator(x, y)
```
